```python
%matplotlib inline
import pandas as pd
import numpy as np
import os

TIME_MATCHING_TOL = 0.1

path = os.getcwd()
dano_eeg = "C:\\Users\christopher.adkins\Desktop\KEEGLogger\KEEGLogger\session_data\dano\PIN_FIXED_4\dano_PIN_FIXED_4_2017-10-17-21-34-53_2017-10-17-21-36-30_EEG.csv"
dano_numbers = "C:\\Users\christopher.adkins\Desktop\KEEGLogger\KEEGLogger\session_data\dano\PIN_FIXED_4\dano_PIN_FIXED_4_2017-10-17-21-34-53_2017-10-17-21-36-30_MRK.csv"

eeg_data = pd.read_csv(dano_eeg)
marker_data = pd.read_csv(dano_numbers)

marker_data['timestamp'] = pd.to_datetime(marker_data['timestamp'],unit='s') - pd.Timedelta('04:00:00')
eeg_data['timestamp'] = pd.to_datetime(eeg_data['timestamp'],unit='ms')
eeg_data = eeg_data[eeg_data['timestamp'] > marker_data['timestamp'][0]-pd.Timedelta(milliseconds=235)]
eeg_data['key marker'] = ''

#df.set_index('timestamp')
#235 ms before stroke
#450 ms after stroke
```

```python
import datetime

n = 0
keyMarkerColumn=[]
for index,row in marker_data.iterrows():
    timeToMatch = row[0]
    key = row[1]
    while(n < len(eeg_data)):
        delta = timeToMatch + pd.Timedelta(milliseconds=450) - eeg_data['timestamp'][n]
            
        d = delta.total_seconds()
        n += 1
        keyMarkerColumn.append(key)
        
        if (d<TIME_MATCHING_TOL):
            break

keyMarkerColumn += ['']* (len(eeg_data)-len(keyMarkerColumn))
eeg_data['key marker'] = keyMarkerColumn
```

```python
data = eeg_data.set_index('timestamp')
y_train = data['key marker'][:20000].values
x_train = data[['TP9','AF7','AF8','TP10']][:20000].values
y_test = data['key marker'][20000:].values
x_test = data[['TP9','AF7','AF8','TP10']][20000:].values
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.layers import Embedding
from keras.layers import LSTM

model = Sequential()
model.add(Dense(32,input_shape=(5,1,4)))
model.add(LSTM(32))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])

model.fit(x_train, y_train, batch_size=150, epochs=10)
score = model.evaluate(x_test, y_test, batch_size=16)
```
